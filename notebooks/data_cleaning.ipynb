```python
import pandas as pd
```

```python
df = pd.read_csv("yahoo_stock_data.csv")
```

```python
df.rename(columns={'Ticker': 'Date'}, inplace=True)
```

```python
df = df[~df['Date'].isin(['Price', 'Date'])].copy()
df.dropna(how='all', inplace=True)
```

```python
price_types = ['Open', 'High', 'Low', 'Close', 'Volume']
new_columns = []

for col in df.columns:
    if col == 'Date':
        new_columns.append('Date')
    elif '.NS' in col:
        # Extract the ticker symbol (e.g., RELIANCE.NS)
        base = col.split('.NS')[0] + '.NS'
        # Extract the suffix (.1, .2, etc.)
        suffix = col.split('.')[-1]
        
        if suffix.isdigit():
            # Match the digit to the price_types list
            idx = int(suffix)
            new_columns.append(f"{price_types[idx]}_{base}")
        else:
            # The column with no digit is the 'Open' price
            new_columns.append(f"{price_types[0]}_{base}")
    else:
        new_columns.append(col)

df.columns = new_columns
```

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', inplace=True)
df.fillna(method='ffill', inplace=True)
```

```python
cols_to_fix = df.columns.drop('Date')
df[cols_to_fix] = df[cols_to_fix].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
stocks = ['HDFCBANK.NS', 'ICICIBANK.NS', 'AXISBANK.NS', 'TCS.NS', 'INFY.NS', 'WIPRO.NS', 'RELIANCE.NS', 
    'ONGC.NS', 'BPCL.NS', 'TMCV.NS', 'LT.NS', 'ULTRACEMCO.NS', 'HINDUNILVR.NS', 'ITC.NS', 'NESTLEIND.NS']

for stock in stocks:
    df[f'Daily_Return_{stock}'] = df[f'Close_{stock}'].pct_change()

for stock in stocks:
    df[f'PnL_{stock}'] = df[f'Daily_Return_{stock}'] * df[f'Close_{stock}'].shift(1)
```

```python
sector_map = {
    "HDFCBANK.NS": "Banking", "ICICIBANK.NS": "Banking", "AXISBANK.NS": "Banking",
    "TCS.NS": "IT", "INFY.NS": "IT", "WIPRO.NS": "IT",
    "RELIANCE.NS": "Energy", "ONGC.NS": "Energy", "BPCL.NS": "Energy",
    "TMCV.NS": "Manufacturing", "LT.NS": "Manufacturing", "ULTRACEMCO.NS": "Manufacturing",
    "HINDUNILVR.NS": "FMCG", "ITC.NS": "FMCG", "NESTLEIND.NS": "FMCG"
}
```

```python
stocks = sorted(
    set(col.replace('Close_', '') for col in df.columns if col.startswith('Close_'))
)
```

```python
long_data = []

for stock in stocks:
    temp = pd.DataFrame({
        'Date': df['Date'],
        'Stock': stock,
        'Sector': sector_map.get(stock),
        'Open': df[f'Open_{stock}'],
        'High': df[f'High_{stock}'],
        'Low': df[f'Low_{stock}'],
        'Close': df[f'Close_{stock}'],
        'Volume': df[f'Volume_{stock}'],
        'Daily_Return': df.get(f'Daily_Return_{stock}'),
        'PnL': df.get(f'PnL_{stock}')
    })
    
    long_data.append(temp)

final_df = pd.concat(long_data, ignore_index=True)
```

```python
final_df.sort_values(['Stock', 'Date'], inplace=True)
final_df.reset_index(drop=True, inplace=True)
```

```python
final_df = final_df.dropna(subset=['Close'])
```

```python
final_df.to_csv("cleaned_yahoo_stock_data.csv", index=False)
```
