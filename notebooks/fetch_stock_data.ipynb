```python
import yfinance as yf
import pandas as pd
```

```python
stocks = [
    # Banking
    "HDFCBANK.NS", "ICICIBANK.NS", "AXISBANK.NS",

    # IT
    "TCS.NS", "INFY.NS", "WIPRO.NS",

    # Energy
    "RELIANCE.NS", "ONGC.NS", "BPCL.NS",

    # Manufacturing 
    "TMCV.NS", "LT.NS", "ULTRACEMCO.NS",

    # FMCG
    "HINDUNILVR.NS", "ITC.NS", "NESTLEIND.NS"
]

data = yf.download(
    tickers=stocks, 
    start="2024-01-01", 
    end="2026-01-01", 
    group_by='ticker',
    session=None 
)
```

```python
data.to_csv("yahoo_stock_data.csv")
```
